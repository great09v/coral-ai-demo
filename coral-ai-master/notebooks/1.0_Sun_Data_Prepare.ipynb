```python
import os

import pandas as pd
import numpy as np
```

```python
PATH = '../data_splits/'
```

```python
val_meta = pd.read_csv(os.path.join(f'{PATH}', 'public_labels_cal.csv'))
train_meta = pd.read_csv(os.path.join(f'{PATH}', 'public_labels_train.csv'))
test_meta = pd.read_csv(os.path.join(f'{PATH}', 'public_labels_test.csv'))
```

```python
def process_csv(data, PATH, tag):

    bucket = {}
    for image_name, image_points in data.groupby('image_name'):
        bucket[image_name] = ' '.join([str(i) for i in image_points['label'].unique()])
    
    result = pd.DataFrame.from_dict(bucket, orient='index', columns=['tags'])
    result.to_csv(os.path.join(f'{PATH}', f'{tag}.csv'))
#     return bucket
```

```python
process_csv(test_meta, PATH, 'test')
process_csv(val_meta, PATH, 'val')
process_csv(train_meta, PATH, 'train')
```

### TIFF to JPG

```python
import os, sys
from PIL import Image

error = []
for infile in os.listdir("../data/train/"):
    print("file : " + infile)
    if infile[-3:] == "tif" or infile[-3:] == "bmp" :
       # print "is tif or bmp"
        outfile = infile[:-3] + "jpeg"
        try:
            im = Image.open("../data/train/" + infile)
            print("new filename : " + outfile)
            out = im.convert("RGB")
            out.save("../data/train/" + outfile, "JPEG", quality=90)
        except:
            error.append(infile)
            print('problem: ' + infile)
```

## Loading the metadata for the data splits

```python
val_meta = pd.read_csv(os.path.join('data_splits', 'public_labels_cal.csv'))
train_meta = pd.read_csv(os.path.join('data_splits', 'public_labels_train.csv'))
test_meta = pd.read_csv(os.path.join('data_splits', 'public_labels_test.csv'))
```

```python
def process_images(data, split=""):
    """This function generates patches from the lsit of images."""
    patch_directory = os.path.join(OUT_DIR, split)
    
    for image_name, image_points in data.groupby('image_name'):
        try:
            im = get_image(os.path.join(IMAGES_DIR, image_name+'.tif'))

        except IOError as e:
            print(str(e))
            logging.debug("Couldn't find image {}".format(image_name))
        else:
            image_points = image_points.set_index(['row', 'col'])
            patches = get_patches_from_image(im, image_points.index, 127)

            if patches is not None:
                
                if not os.path.exists(patch_directory):
                    os.makedirs(patch_directory)

                write_patches_as_images(image_name, patches=patches, 
                                        labels=image_points.label, 
                                        out_dir=patch_directory)
```

```python
process_images(train_meta, split='train')
```

```python
process_images(val_meta, split="validation")
```

```python
process_images(test_meta, split='test')
```
